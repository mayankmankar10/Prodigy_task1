```python
import pandas as pd
df = pd.read_csv("student-por.csv")
df
```

```python
X = df.head(100)
X
```

```python
import matplotlib.pyplot as plt
#Continous data - age
#Histogram of distribution of students with ages.
plt.hist(df['age'],bins = 20, color = 'brown', edgecolor = 'black')
plt.xlabel('Age')
plt.ylabel('No. of Students')
plt.title('Histogram for distibution for ages of students.')
plt.show()
```

```python
#Histogram of Student data set showing variation of score with age.
age = X['age']
Grade_3 = X['G3']
plt.bar(Grade_3, age, color = 'lightblue')
plt.title('Variation of scored grade 3 with age')
plt.xlabel('Age')
plt.ylabel('Score')
plt.show()
```

```python
#Categorical data
#Histogram of the number of males and females.
plt.hist(df['sex'], bins = 3, color = 'blue')
plt.title('No. of males and females in school.')
plt.xlabel('Gender/sex')
plt.ylabel('No. of students')

plt.show()
```
